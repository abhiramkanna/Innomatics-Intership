```python
#Internship Logic Building Task – 7
#1.Smart Light Controller
#Code
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def switch_on(self):
        self.status = "ON"

    def show_status(self):
        print(self.light_name, "is", self.status)


# Input
light = SmartLight("Bedroom Light")
light.switch_on()
light.show_status()
```

```python
#2.Employee ID Card System
#Code
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


emp = Employee("Rahul", "EMP102", "IT")
emp.display_id()
```

```python
#3.Mobile Contact Record
#Code
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c1 = Contact("Anita", "9876543210")
c1.display_contact()
```

```python
#4.Product Price Tag Generator
#Code
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))


p = Product("Headphones", 2499)
p.price_tag()
```

```python
#5.Movie Rating Display System
#Code
class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


m = Movie("Inception", 4.8)
m.display_movie()
```

```python
#6.Delivery Address Manager
#Code
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


d = Delivery("Suman", "Hyderabad")
d.delivery_details()
```
