```python
#Python Programming Assignment Data Processing and Analysis (Task-3)
# 1.Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# find highest score
highest = max(employees.values())

top = []
for name in employees:
    if employees[name] == highest:
        top.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top), "(Score:", highest, ")")
print()
```

```python
# 2.Search Query Keyword Analysis
text = "Buy mobile phone buy phone online"

# convert to lowercase
text = text.lower()

# remove punctuation (simple way)
text = text.replace(",", "").replace(".", "")

# split into words
words = text.split()

# count frequency
count = {}
for w in words:
    count[w] = count.get(w, 0) + 1

# print only words with count > 1
result = {}
for k in count:
    if count[k] > 1:
        result[k] = count[k]
print(result)
```

```python
# 3.Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
#condition vails readings
valid = []
#loop 
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:   # check even
        valid.append((i, sensor_readings[i]))
#print the readings
print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

```python
# 4.Email Domain Usage Analysis

emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count = {}

# count domains
for e in emails:
    domain = e.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

# calculate percentage
for d in domain_count:
    percent = (domain_count[d] / total) * 100
    print(f"{d}: {int(percent)}%")
```

```python
# 5.Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

# average sales
avg = sum(sales) / len(sales)

# check for >30% above average
limit = avg * 1.3

for i in range(len(sales)):
    if sales[i] > limit:
        print(f"Day {i+1}: {sales[i]}")
```

```python
# 6. Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# count each id
for u in user_ids:
    count[u] = count.get(u, 0) + 1

# print only duplicates
for u in count:
    if count[u] > 1:
        print(u, "→", count[u], "times")
```
