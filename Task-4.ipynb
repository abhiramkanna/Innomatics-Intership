```python
#Task 4: Feb Internship Function with Loops
# 1.Mobile Recharge Validation System
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Amount too low"
    if plan not in valid_plans:
        return "Invalid plan"
    return "Recharge successful"


while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan: "))

    result = validate_recharge(amount, plan)
    print(result)

    if result == "Recharge successful":
        break
    else:
        print("Please try again\n")
```

```python
# 2.Inventory Reorder Alert System
#function
def check_stock(products):
    for item, qty in products.items():
        if qty < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")
#dictionary values

inventory = {
    "Pen": 20,
    "Notebook": 10,
    "combox": 55,
    "Eraser": 25
}

check_stock(inventory)
```

```python
# 3.Student Result Processing System
#function
def process_result(marks):
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)

    print("Average:", avg)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"
#list of students

marks_list = [50, 45, 70, 55]
result = process_result(marks_list)
print("Result:", result)
```

```python
# 4.Cab Fare Estimator with Retry Option
#function
def calculate_fare(km, peak):
    fare = 50 + (12 * km)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare

#while condition for input
while True:
    km = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ")

    total = calculate_fare(km, peak)
    print("Estimated Fare:", total)

    retry = input("Retry? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# 5.Employee Attendance Eligibility Checker
# function
def check_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1
# formula of percentage
    percent = (present / len(attendance)) * 100
    print("Attendance %:", percent)

    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

#sample list
att_list = ["P", "P", "A", "P", "P", "P", "A", "A"]
print(check_eligibility(att_list))
```

```python
# 6.Password Strength Checker
# function
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    # check minimum length
    if len(password) < 8:
        return "Weak password"

    # loop through characters
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # final condition check
    if has_digit and has_special:
        return "Strong password"
    else:
        return "Weak password"


pwd = input("Enter password: ")
print(check_password(pwd))
```
