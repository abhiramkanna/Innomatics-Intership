Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total=0
    for i in likes_list:
        total+=i

    if total>=1000:
        return "Viral Post"
    else:
        return "Normal Engagement"


likes_list=[150,554,231,566,885]
result=analyze_post_engagement(likes_list)
print(result)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_adequacy_checker(rainfall_data,required_level):
    sum=0
    for i in rainfall_data:
        sum+=i
    average=sum//len(rainfall_data)

    if average>required_level:
        status="Adequate Rainfall"

    else:
        status="Inadequate Rainfall"


    print(f"Average Rainfall:{average}")
    print(f"Rainfall Status:{status}")

rainfall_data=[56,48,55,88,99]
required_level=70

rainfall_adequacy_checker(rainfall_data,required_level)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def duplicate_account_detection(account_list):
    count = {}

    for i in account_list:
        if i in count:
            count[i] += 1
        else:
            count[i] = 1

    for user in count:
        if count[user] > 1:
            print("Duplicate Accounts Found: Yes")
            return

    print("Duplicate Accounts Found: No")


account_list = ["tarak", "raj", "sai", "tarak"]
duplicate_account_detection(account_list)
```

```python
def application_health_checker(errors):
    if errors==0:
        status="Healthy"
    elif errors<=5:
        status="Minor Issues"
    elif errors>5:
        status="Critical Issues"


    print(f"Error Count:{errors}")
    print(f"System Status:{status}")

application_health_checker(5)
```

9: E-Learning – Student Attendance Eligibility System

```python
def student_attendance_eligibility(attendance_data):
    sum=0
    for i in attendance_data:
        if i=="P":
          sum+=1
    average=(sum/len(attendance_data))*100
    if average>=75:
        status="Eligible"
    else:
        status=" Not Eligible"
    print(f"Attendance Percentage:{average}")
    print(f"Exam Eligibility:{status}")
attendance_data=["P","P","A","A","P"]
student_attendance_eligibility(attendance_data)
```

Problem 10: Smart Electricity Bill Analyzer

```python
def smart_electricity_bill_analyzer(units):
    #billing rules
    if units<=100:
        total_bill=units*3
    elif units>100 and units<=200:
        total_bill=units*5
    elif units>200:
        total_bill=units*7
    #usage classification
    if total_bill<500:
        status="Low Usage"
    elif total_bill>500 and total_bill<=1500:
        status="Moderate Usage"
    elif total_bill>1500:
        status="High Usage"

    print(f"total_bill:{total_bill}")
    print(f"Usage:{status}")
        
smart_electricity_bill_analyzer(300)
```
